```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p(x):
  return 1/(2* (x+1))

def q(x):
  return -1

def f_func(x):
  return -np.sqrt(x+1)
```

```python
def diff_scheme(h, pi):
  r = pi/2 * h
  ai = 1/(h**2) * (1 + r**2/2 - r)
  ci = 1/(h**2) * (1 + r**2/2 + r)
  return ai, ci
```

```python
N = 3
xn, x0 = 1, 0
h = (xn - x0) / N
x = np.linspace(x0, xn, N+1)

A = np.zeros((N+1, N+1))
f = np.zeros(N+1)

for i in range(1, N):
    pi = p(x[i])
    ai, ci = diff_scheme(h, pi)
    bi = q(x[i]) - ai - ci

    A[i, i-1] = ai
    A[i, i] = bi
    A[i, i+1] = ci

    f[i] = f_func(x[i])

A[N, N] = 1
f[N] = np.sqrt(2)

A[0, 0] = -3 /(2*h)
A[0, 1] = 4 / (2*h)
A[0, 2] = -1 / (2*h)

A[N, N] = 1
f[0] = 0.5

u_exact = np.sqrt(x + 1)
u = np.linalg.solve(A, f)
```

```python
print(np.max(np.abs(u - u_exact)))

plt.plot(x, u, label='Численное решение', linestyle='--')
plt.plot(x, u_exact, label='Аналитическое решение')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.grid()
plt.show()
```
