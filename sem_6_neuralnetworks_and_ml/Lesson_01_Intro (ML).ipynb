````python
# создадим функцию, возвращающую квадрат числа
def x2(x): return x*x

# Обозначим иксы в промежутке от -4 до 4
xs = list(range(-4, 5))

# рассчитаем значения y
ys = [ x2(x) for x in xs ]

# отрисовывать можно и просто вызвав функцию plot
plt.plot(xs, ys)

plt.show()
````

Перемножение матриц


````python
# Создадим по две матрицы размером 100 на 100 на питоне и в numpy

n_rows = 100
n_cols = 100

python_A = [ list(range(n_cols)) for _ in range(n_rows) ]
python_B = [ list(range(n_cols)) for _ in range(n_rows) ]

np_A = np.array([ np.array(r) for r in python_A ])
np_B = np.array([ np.array(r) for r in python_B ])
````

````python
# Медленная кастомная функция перемножения двух матриц (не надо так)
def custom_mat_mul(A, B):
  n = len(A)
  k1 = len(A[0])
  k2 = len(B)
  m = len(B[0])

  if k1 != k2:
    raise ValueError(f'Multiplying matrix with shape {n}x{k1} and matrix with shape {k1}x{m}. {k1} != {k2}')

  k = k1

  ans = [ [0] * n for _ in range(m) ]
  for t in range(m):
    for i in range(n):
      v = 0
      for j in range(k):
        v += A[i][j] * B[j][t]
      ans[i][t] = v

  return ans
````

````python
# Проверим, что ответ нашей функции совпадает с numpy
assert (np_A @ np_B == custom_mat_mul(python_A, python_B)).all() == True

# операнд @ озачает матричное перемножение, также явно можно использовать np.matmul
assert (np.matmul(np_A, np_B) == np_A @ np_B).all() == True
````

````python
%timeit -n 25 custom_mat_mul(python_A, python_B)
````

````python
%timeit -n 25 np_A @ np_B
````

В этом случае разница уже в **100** раз.


### Почему numpy быстрый

В отличие от питоновских листов, которые могут содержать разнородные объекты - из-за чего они медленные, ndarray массивы numpy однородны, то есть способны содержать в себе только данные одного типа, фиксированной длины. Это позволяет хранить все данные **последовательно** в памяти.

Это способствует лучшему кэшированию CPU - что сильно ускоряет вычисления.

Так же numpy разбивает большие операции на множество маленьких и параллелизирует их с помощью SIMD (SSE, AVX).

### 1) corrcoef
Реализовать функцию расчета [коэффицента корреляции Пирсона](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D1%80%D1%80%D0%B5%D0%BB%D1%8F%D1%86%D0%B8%D1%8F#%D0%9B%D0%B8%D0%BD%D0%B5%D0%B9%D0%BD%D1%8B%D0%B9_%D0%BA%D0%BE%D1%8D%D1%84%D1%84%D0%B8%D1%86%D0%B8%D0%B5%D0%BD%D1%82_%D0%BA%D0%BE%D1%80%D1%80%D0%B5%D0%BB%D1%8F%D1%86%D0%B8%D0%B8) для двух numpy векторов с помощью функционала numpy (без циклов)

P.S. нельзя использовать функцию np.corrcoef

````python
def corrcoef(v1, v2):
  E1 = np.mean(v1)
  E2 = np.mean(v2)
  return np.mean((v1 - E1) * (v2 - E2)) / (np.std(v1) * np.std(v2))
````

````python
# Test 1
v1 = np.random.rand(100)
v2 = np.random.rand(100)
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

````python
# Test 2
v1 = np.array([1, 2, 3])
v2 = np.array([1, 2, 3])
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

````python
# Test 3
v1 = np.random.normal(loc=0, scale=1, size=100)
v2 = v1 + 0.2 * np.random.rand(100)
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

### 3) dist
Реализовать функцию dist(v1, v2), которая принимает на вход два вектора с одинаковым shape, и возвращает число - [евклидово расстояние](https://ru.wikipedia.org/wiki/Евклидова_метрика) между ними.

Так же нельзя использовать циклы и np.linagl.norm

````python
def dist(v1, v2):
  return np.sqrt(np.sum((v1 - v2)**2))
````

````python
# Test 1
v1 = np.array([0, 0, 1])
v2 = np.array([0, 0, 0])

assert dist(v1, v2).round(8) == np.linalg.norm(v1-v2).round(8)
````

````python
# Test 2
v1 = np.array([2, 3, 1])
v2 = np.array([0, 0, 0])

assert dist(v1, v2).round(8) == np.linalg.norm(v1-v2).round(8)
````

````python
# Test 3
v1 = np.random.rand(100)
v2 = np.random.rand(100)

assert dist(v1, v2).round(8) == np.linalg.norm(v1-v2).round(8)
````
