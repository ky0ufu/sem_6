```python
import numpy as np
import matplotlib.pyplot as plt


N = 10
K = 10
ua = 0
ub = 1
T = 1.0
h = (ub - ua) / N
tau = T / K
a = 1
r = a * tau / h
print(f"r = {r}")


def g(x, t):
  return 0

def phi(x):
  return 0

def ksi(x):
  return 0

def gamma0(t):
  return -3*t**2

def gamma1(t):
  return -2*t


x = np.linspace(ua, ub, N+1)
t = np.linspace(0, T, K+1)

u = np.zeros((K+1, N+1))


u[0, :] = phi(x)

for n in range(1, N):
  u[1, n] = u[0, n] + r**2 * (u[0, n-1] - 2*u[0, n] + u[0, n+1]) + tau*u[0, n]


u[:, 0] = gamma0(t)
u[:, -1] = gamma1(t)

for k in range(1, K):
    for i in range(1, N):
        u[k+1, i] = 2 * (1 - r**2) * u[k, i] - u[k-1, i] + r**2 * (u[k, i+1] + u[k, i-1]) + tau**2*g(x[i], t[k])

```
